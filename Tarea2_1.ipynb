```python
#Predicción del peso con las características de altura y peso.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./peso.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.hist(bins=50, figsize=(20,15), edgecolor='black')
```

```python
import seaborn as sb
```

```python
sb.scatterplot(data=df, x='Altura', y='Peso', hue='Peso', palette='coolwarm')
```

```python
X = df[['Altura']]
y = df['Peso']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression

modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
predicciones = modelo.predict(X_test)
```

```python
modelo.predict([[1.54]])
```

```python
modelo.score(X_test, y_test)
```

```python
modelo.score(X_train, y_train)
```

```python
print(f"Peso: {modelo.coef_}")
print(f"sesgo (Bias): {modelo.intercept_}")
```

```python
#El modelo únicamente con las características de altura y peso funciona, pero no de la mejor manera,
#ya que el porcentaje de los datos de prueba fue de 48%, y en los datos de entrenamiento da un 51%,
#esto puede significar que el modelo no esta capturando completamente las relaciones en los datos
#ya que como se observa en la gráfica la relación entre las variables no es completamente lineal y
#existen mas características importantes para determinar el peso de una persona, lo cual ayudaria
#al modelo a mejorar la predicción.
```

```python
df2 = pd.read_csv('./peso2.csv')
df2.head()
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2.hist(bins=50, figsize=(20,15), edgecolor='black')
```

```python
sb.scatterplot( data=df2, x='Altura', y='Peso', hue='Edad', palette='coolwarm')
```

```python
X2 = df2[['Altura', 'Edad']]
y2 = df2['Peso']
```

```python
from sklearn.model_selection import train_test_split

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression

modelo2 = LinearRegression()
modelo2.fit(X2_train, y2_train)
```

```python
predicciones2 = modelo2.predict(X2_test)
```

```python
modelo2.predict([[1.54, 24]])
```

```python
modelo2.score(X2_test, y2_test)
```

```python
modelo2.score(X2_train, y2_train)
```

```python
print(f"Peso: {modelo2.coef_}")
print(f"sesgo (Bias): {modelo2.intercept_}")
```

```python
#El modelo agregando la caracteristica de edad funciona mejor en comparacion al ejemplo anterio,
#pero aun le falta por mejorar, ya que el porcentaje de los datos de prueba fue de 89%, y en los
#datos de entrenamiento da un 69%, de igual manera se puede dar porque el modelo no esta capturando
#completamente las relaciones en los datos y en la gráfica observamos un comportamiento parecido al
#anterior donde la relación no es completamente lineal y tambien existen mas características importantes
#para determinar el peso de una persona, y la muestra de datos es pequeña.
```
